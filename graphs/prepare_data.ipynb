```python
# importa data

import pandas as pd
import pickle
import os
import glob

os.chdir("../results")
```

```python
# list all files from results folder
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]

# sort all_filenames
all_filenames.sort()
```

```python
import csv
import ast

def read_csv_to_dict(file_path, flag=False):
    placement_dict = {}
    
    with open(file_path, mode='r') as file:
        csv_reader = csv.reader(file)
        for row in csv_reader:
            app_name = row[0]  # The first element (e.g., 'App4')
            # Convert the string representation to actual Python objects
            values_str = row[1]
            parsed_values = ast.literal_eval(values_str)
            
            # Extract values from the tuple
            if flag:
                tuple_list = parsed_values[0]  # The list of tuples
                float_value_str = [str(parsed_values[1])]  # The floating-point value as a string inside a list
                extra_float_value = [str(parsed_values[2])]   # Additional float value
                status_str = parsed_values[3]  # Additional string value ('Optimal')
                placement_dict[app_name] = {
                    "value1": tuple_list, 
                    "value2": float_value_str,
                    "value3": extra_float_value,
                    "status": status_str
            }
            else:
                tuple_list = parsed_values[0]  # The list of tuples
                float_value_str = [str(parsed_values[1])]  # The floating-point value as a string inside a list
                # Store all values in the dictionary
                placement_dict[app_name] = {
                    "value1": tuple_list, 
                    "value2": float_value_str
                }
            
    return placement_dict

# prepare data
def prepare_data(data, flag=False):
    allocations = dict()
    times = dict()
    for key, value in data.items():
        allocations[key]= value['value1']
        if flag:
            times[key] = value['value3']
        else:
            times[key] = value['value2']

    for key, value in allocations.items():
        for i in range(len(value)):
            value[i] = (value[i][1], value[i][0])

    times = {key: [value[0].strip("[]'")] for key, value in times.items()}

    return allocations, times
```

```python
# Example usage:
for file_path in all_filenames:

    print(file_path)
    if file_path.startswith('m'):
        csv_data = read_csv_to_dict(file_path, True)
        allocations, times_p = prepare_data(csv_data, True)
    else:
        csv_data = read_csv_to_dict(file_path)
        allocations, times_p = prepare_data(csv_data)

    # h1_min_nodes_alloc.csv take off .cvs
    new_file_path = file_path[:-4]

    # print(csv_data)
    print(allocations)
    print(times_p)
    with open('placements/'+new_file_path+'_place.pkl', 'wb') as f:
        pickle.dump(allocations, f)

    with open('times/'+new_file_path+'_time.pkl', 'wb') as f:
        pickle.dump(times_p, f)

    print('File written!!!')
    print('_______________')

```
